# DATA CLEANING


```python
# looking for duplicates in the dataset 

data.duplicated().sum()
```

```python
# finding out how many null values are in the dataset

data.isnull().sum()
```

```python
# filling the null values in the certificate column with what was in the previous row
new_data = data.copy()
new_data['Certificate'] = data['Certificate'].ffill()	
new_data
```

```python
# fill the null values in the meta_score column with the mean

new_data['Meta_score'] = data['Meta_score'].fillna(data['Meta_score'].mean())
print(new_data)
```

```python
# finding out the data type of each column

print(new_data.dtypes)
```

```python
# filling missing values with median of the column

new_data['Gross'] = new_data['Gross'].replace(',', '', regex=True).astype(float) # change data type to numeric

new_data['Gross'] = new_data['Gross'].fillna(new_data['Gross'].median())

new_data
```

```python
new_data.isnull().sum()
```

```python
print(new_data.columns) 
```

# DATA VISUALISATION

```python
#visualising the distribution of ratings in a box plot

new_data.IMDB_Rating.value_counts().plot(kind = 'box')
```

```python
# visualising the distribution of ratings in bargarph

new_data.IMDB_Rating.value_counts().plot(kind = 'bar')
```

```python
# visualising the distribution of ratings using a histogram 

plt.hist(new_data['IMDB_Rating'], bins=20, edgecolor = 'grey', color = plt.cm.viridis(0.2))

plt.xlabel('IMDB_Rating')
plt.ylabel('frequency')
plt.title('The Distribution of IMDB_Rating')

plt.show()
```

```python
# Identifying the top five rated movies and genres

top_rated_movies = new_data.nlargest(8, 'IMDB_Rating')[['Genre', 'IMDB_Rating']]
top_rated_movies
```

```python
# visualising the findings

sns.barplot(x='Genre', y='IMDB_Rating',hue='IMDB_Rating', data = top_rated_movies, palette="viridis", legend=False)
plt.title("IMDB Ratings by Genre", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

```python
#visualising the correlation between ratings and genre

sns.scatterplot(x='Genre', y='IMDB_Rating', data = top_rated_movies)
plt.title("IMDB Ratings by Genre", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```
