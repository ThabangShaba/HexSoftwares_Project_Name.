# Cleaning the data
 

```python
data.isnull().sum()
```

```python
data.drop_duplicates()
```

```python
# dropping unwanted columns
data_dp = data.drop(columns= ['Embarked', 'Cabin'])
data_dp
```

```python
# filling missing values with the mean

data_dp['Age'] = data_dp['Age'].fillna(data_dp['Age'].mean()).round().astype(int)
```

```python
data_dp
```

```python
data_dp.isnull().sum()
```

```python
# changing format in name column to initial, first name then last name

def format_name(name):
    parts = name.split()
    
    if len(parts) >= 2:
        initials = "".join([p[0].upper() + "." for p in parts[:-1]])  
        first_name = parts[0].title()  
        last_name = parts[-1].title()  
        return f"{initials} {first_name} {last_name}"
    return name  


data_dp['Name'] = data_dp['Name'].apply(format_name)


data_dp
```

```python
# Remove full stops inside brackets
data_dp['Name'] = data_dp['Name'].str.replace(r'\((.*?)\.', r'(\1', regex=True)
data_dp
```

```python
# removing brackets in name column

data_dp['Name'] = data_dp['Name'].str.replace(r'\(.*\)', '', regex=True)
data_dp
```

```python
# Adding commas where its missing in name column

def add_comma_if_missing(name):
    if ',' not in name:  
        parts = name.split()
        if len(parts) >= 2:
            return f"{parts[0]}, {parts[-1]}"  
    return name  

data_dp['Name'] = data_dp['Name'].apply(add_comma_if_missing)
data_dp
```

```python
#check the type of errors or issues in a column
def check_errors_in_column(column):
    errors = {
        'NaN values': column.isna().sum(),
        'Non-string values': column[~column.apply(lambda x: isinstance(x, str))].count(),
        'Empty strings': (column == '').sum(),
        'Invalid numbers': column.apply(lambda x: isinstance(x, str) and x.replace('.', '', 1).isdigit()).sum(),
    }
    return errors

data_errors = check_errors_in_column(data_dp['Fare'])

print(data_errors)
```

```python
# filling missinng values in fare variables with the median

data_dp['Fare']= data_dp['Fare'].fillna(data_dp['Fare'].median())
print(data_dp)
```
